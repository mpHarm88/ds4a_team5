### Subesetting PPP Data based on Zip codes pre covid

```python
yelp_business_path = "../data/yelp_dataset/yelp_academic_dataset_business.json"

yelp_b_df = pd.read_json(yelp_business_path, lines=True)
under_150 = pd.read_csv("../data/under150k.csv")

print(yelp_b_df.shape)
```

```python
yelp_b_df.head()
```

```python
def zip_subsetter(df, df2):
    
    """
    This function will take in ppp data or a subset of ppp data 
    and filter the data based on zipcodes found in the 'yelp academic dataset business'
    
    df = ppp data to subset
    df2 = yelp_academic_dataset_business read in as a pandas DataFrame
    """
    
    # ppp data
    df = df.copy()
    
    # yelp_academic_dataset_business
    df2 = df2.copy()
    
    # PPP zip code cleaning
    # Filter Zip codes for rows that dont contain NaN
    # Cast as a String data type
    # Apply function that removes the decimal value by subsetting string
    df["Zip"] = df[df["Zip"].isna() == False].astype(str)["Zip"].apply(lambda x: x[:5])

    # Filter Zip codes thata are not NaN values generated after cleaning
    df = df[df["Zip"].isna() == False]
    
    # List of yelp zipcodes to subset ppp data with
    yelp_zips = list(df2["postal_code"])
    
    # Subset ppp data based on zipcodes present in cleansed yelp zip code data
    ppp_zip_subset = df[df["Zip"].isin(yelp_zips)]
    
    print(f"Shape of ppp_zip_subset: {ppp_zip_subset.shape}")
    
    return df
```

```python
df = zip_subsetter(under_150, yelp_b_df)
df.head()
```
